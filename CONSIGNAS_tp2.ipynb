### 1. **Bucles**

   - **Actividad 1.1**: Escribir un programa que muestre el print de todo lo que el usuario introduzca hasta que el usuario escriba “salir” que terminará.
   
   - **Actividad 1.2**: Escribir un programa que pida al usuario un número entero y muestre por pantalla un triángulo rectángulo como el de más abajo con la cantidad de escalones indicados previamente (los numeros dentro del triangulo deben ser enteros aleatorios del 0 al 9).
* 1
* 3 1
* 5 3 1
* 7 5 3 1

```python

# todo: Resolución
# ! 1

def imprimir(user_input):
    print(user_input)

while True:
        user_input = input("Escribe algo o 'salir' para terminar: ")
        
        if user_input == "salir":
            break
        
        imprimir(user_input)
```

```python

# todo: resolución

# ! 2

import random 

def triangulo(escalones):
    # Generar cada fila del triángulo
    for i in range(1, escalones + 1):
        fila = ""
        # Agregar números aleatorios a la fila
        for _ in range(i):
            fila += str(random.randint(0, 9))
        print(fila.strip())  # Imprimir la fila generada

# Solicitar al usuario la cantidad de escalones
escalones = int(input("Ingrese la cantidad de escalones: "))
triangulo(escalones)
```

### 2. **Listas y Tuplas**

   - **Actividad 1**: Escribir un programa que pida al usuario una frase y muestre por pantalla el número de veces que contiene cada vocal y la cantidad de espacios en blanco.

   - **Actividad 2**: Escribir un programa que cree una matriz (con numeros aleatorios del 1 al 100) usando las dimensiones de X e Y que el usuario deberia ingresar por consola, luego crea una tupla con la matriz transpuesta y por ultimo las muestra ambas por consola .

   - **Actividad 3**: Escribir un programa que almacene en una lista los siguientes precios, 50, 75, 46, 22, 80, 65, 8, y muestre por pantalla el menor y el mayor de los precios

```python

# todo: Resolución

# ! 1

def contar_vocales(frase):
    # Inicializar el diccionario para contar vocales y espacios
    conteo = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0, "espacios": 0}

    # Recorrer cada letra de la frase en minúsculas
    for letra in frase.lower():
        if letra in conteo:  # Si es una vocal, incrementar su conteo
            conteo[letra] += 1 
        elif letra == ' ':  # Si es un espacio, incrementar el conteo de espacios
            conteo['espacios'] += 1

    return conteo

# Solicitar al usuario que escriba una frase
frase = input("Escriba una frase")
resultado = contar_vocales(frase)

# Mostrar los resultados
print(f"La frase es: {frase}")
print("Resultados:")
for vocal, cantidad in resultado.items():
    print(f"{vocal}: {cantidad}")
```

```python

# todo: Resolución

# ! 2

import random

def crear_matriz(columnas, filas):
    matriz = []
    
    # Crear filas con números aleatorios y añadirlas a la matriz
    for _ in range(filas):
        fila = [random.randint(1, 100) for _ in range(columnas)]
        matriz.append(fila)

    # Crear la matriz transpuesta como una tupla
    matriz_transpuesta = tuple(
        tuple(matriz[i][j] for i in range(filas))
        for j in range(columnas)
    )

    # Imprimir la matriz original
    print(f"Matriz {columnas}x{filas}:")
    for filas in matriz:
        print(filas)

    # Imprimir la matriz transpuesta
    print("\nMatriz transpuesta")
    for filas in matriz_transpuesta:
        print(filas)

# Se solicita las dimensiones de la matriz al usuario
columnas = int(input("Ingrese la cantidad de columnas de la matriz: "))
filas = int(input("Ingrese la cantidad de filas de la matriz: "))

# Crear y mostrar la matriz
matriz = crear_matriz(columnas, filas)
```

```python

# todo: Resolución

# ! 3

def menor_mayor(lista_numeros):
    # Encontrar el mayor y menor número en la lista
    mayor = max(lista_numeros)
    menor = min(lista_numeros)
    return mayor, menor

# Lista de precios
precios = [50, 75, 46, 22, 80, 65, 8]

# Obtener el mayor y menor precio
mayor, menor = menor_mayor(precios)

# Mostrar los resultados
print(f"{mayor} es el mayor de los precios")
print(f"{menor} es el menor de los precios")
```

### 3. **Diccionarios**

   - **Actividad 1**: Escribir un programa que cree un diccionario simulando una cesta de la compra. El programa debe preguntar el artículo y su precio y añadir el par al diccionario, hasta que el usuario decida terminar. Después se debe mostrar por pantalla la lista de la compra y el coste total, con el siguiente formato.

   | arts  |precio |
   |-------|-------|
   |art 1  |1      |
   |art 2  |2      |
   |art 3  |3      |
   |total  |6      |

   - **Actividad 2**: Escribir un programa que cree un diccionario de traducción español-inglés. El usuario introducirá las palabras en español e inglés separadas por dos puntos, y cada par <palabra> : <traducción> separados por comas (es decir, solo se debe usar **UNA VEZ** la función input para la carga de palabras). El programa debe crear un diccionario con las palabras y sus traducciones. Después pedirá una frase en español y utilizará el diccionario para traducirla palabra a palabra. Si una palabra no está en el diccionario debe dejarla sin traducir.

   - **Actividad 3**: Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.

```python

# todo: Resolución

# ! 1

def ticket(compras):
    total = 0
    print("\nCesta de compra")
    # Iterar sobre los artículos y precios en el diccionario
    for articulo, precio in compras.items():
        print(f"{articulo.title()}  ${precio}")
        total += precio  # Sumar el precio al total
    print(f"Total: ${total}")

def cargar_compras():
    cesta_compras = {}

    while True:
        # Solicitar el nombre y precio del artículo
        articulo = input("Ingrese en nombre del artículo: ")
        precio = float(input(f"Ingrese el precio de {articulo}"))
        cesta_compras[articulo] = precio  # Agregar artículo a la cesta

        # Verificar si el usuario desea terminar
        print("Ingrese cualquier valor para agregar otro artículo o 1 para terminar")
        opcion = input("Ingrese su opción")

        if opcion == "1":
            break
        
    return cesta_compras

# Cargar los artículos en la cesta y generar el ticket
compras = cargar_compras()
ticket(compras)
```

```python

# todo: Resolución

# ! 2

def crear_diccionario():
    # Solicitar al usuario pares de palabras español:inglés
    entrada = input("Ingrese pares español:inglés separados por comas")
    diccionario = {}
    pares = entrada.split(",")

    for par in pares:
        if ":" in par:  # Verificar que el par contenga ":"
            esp, ing = par.strip().split(":")
            diccionario[esp.strip()] = ing.strip()
    
    return diccionario

def traducir_frase(diccionario):
    # Solicitar al usuario una frase en español
    frase = input("Ingrese una frase en español para traducir")
    palabras = frase.split()
    traduccion = []

    for palabra in palabras:
        # Traducir palabra o dejarla igual si no está en el diccionario
        traduccion.append(diccionario.get(palabra, palabra))

    # Mostrar la frase traducida
    print("Frase traducida:")
    print(' '.join(traduccion))
    

diccionario = crear_diccionario()
traducir_frase(diccionario)
```

```python

# todo: Resolución

# ! 3

def contar_palabra(frase):
    # Crear un diccionario con cada palabra y su frecuencia en la frase
    resultado = {palabra: frase.split().count(palabra) for palabra in frase.split()}
    return resultado

# Solicitar al usuario que ingrese una frase
frase = input("Ingrese una frase: ")

# Obtener el diccionario con las palabras y sus frecuencias
resultado = contar_palabra(frase)

# Mostrar el resultado
print(resultado)
```

### 4. **Funciones**

- **Actividad 1**: Escribir una función que calcule el total de una factura tras aplicarle el IVA. La función debe recibir la cantidad sin IVA y el porcentaje de IVA a aplicar, y devolver el total de la factura. Si se invoca la función sin pasarle el porcentaje de IVA, deberá aplicar un 21%.

- **Actividad 2**: Escribir una función que reciba un número entero positivo y devuelva su factorial.


```python

# todo: Resolución

# ! 1

def calcular_iva(precio, iva=21):
    # Calcular el precio total con IVA
    return precio * (1 + iva / 100)

# Solicitar al usuario el precio sin IVA
precio = float(input("Precio sin IVA: "))

# Solicitar el porcentaje de IVA, con un valor predeterminado de 21%
iva = float(input("IVA(%): ") or 21)

# Mostrar el total con IVA
print(f"Total con IVA ${calcular_iva(precio, iva)}")
```

```python

# todo: Resolución

# ! 2

def calcular_factorial(numero):
    # Inicializar el total en 1
    total = 1
    for i in range(1, numero + 1):
        total *= i  # Multiplicar el total por el número actual
    return total

# Usar la variable 'escalones' como entrada para calcular el factorial
print(f"El factorial de {escalones} es: {calcular_factorial(escalones)}")
```
